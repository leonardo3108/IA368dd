```python
import time
import pickle

def create_chat_messages(instruction_type, texts, texts_shots, sentiments_shots, model_name, output, start=0):
    responses = []
    if instruction_type in ['zero_shot', 'few_shot_with_instruction']:
        messages=[{"role": "system", "content": "You are a helpful assistant that provide sentiment analysis and returns positive, or negative."}]
    else:
        messages = []
    if instruction_type in ['few_shot_with_instruction', 'few_shot_without_instruction']:
        for shot_index in range(len(texts_shots)):
            messages.append({"role": "user", "content": f"{texts_shots[shot_index]}"})
            messages.append({"role": "system", "content": f"{sentiments_shots[shot_index]}"})
    total = len(texts)
    for i, text in enumerate(texts[start:]):
        instance_messages = messages.copy()
        instance_messages.append({"role": "user", "content": f"{text[:2000]}"})
        if i > 0:
            if i % 10 == 0:
              time.sleep(25)
        else:
            print('message 0:')
            for message in messages:
                print('\t', message)
        response = openai.ChatCompletion.create(
            model=model_name,
            messages=instance_messages,
            temperature=0
        )
        pickle.dump(response, output)
        print(f'Processed text {(start + i)} / {total}.')
```

## Retrieving results

```python
def load_responses(file_name):
    responses = []
    with open(file_name, 'rb') as file:
        try: 
            while True:
                responses.append(pickle.load(file))
        except:
            pass
    return responses            
```

```python
responses_zero_shot = load_responses('zero-shot.pkl')
responses_few_shot = load_responses('few-shot-instruction.pkl')
responses_few_shot_noins = load_responses('few-shot-no-instruction.pkl')

print('zero shot:', len(responses_zero_shot))
print('few shot, with instruction:', len(responses_few_shot))
print('few shot, no instruction', len(responses_few_shot_noins))
```

## Extracting messages

```python
def extract_messages(responses):
    return [response.choices[0].message.content for response in responses]

messages_zero_shot = extract_messages(responses_zero_shot)
messages_few_shot = extract_messages(responses_few_shot)
messages_few_shot_noins = extract_messages(responses_few_shot_noins)

print(messages_zero_shot[120])
print(messages_few_shot[120])
print(messages_few_shot_noins[120])
```

## Interpreting sentiment

```python
def extract_sentiments(messages):
    predictions = []
    for message in messages:
        message = message.lower()
        pos = message.index('ositive') if 'ositive' in message.lower() else 9999
        neg = message.index('egative') if 'egative' in message.lower() else 9999
        if pos < neg:
            predictions.append('positive')
        elif neg < pos:
            predictions.append('negative')
        else:
            predictions.append('undefined')
    return predictions

predictions_zero_shot = extract_sentiments(messages_zero_shot)
predictions_few_shot = extract_sentiments(messages_few_shot)
predictions_few_shot_noins = extract_sentiments(messages_few_shot_noins)

print(predictions_zero_shot[120])
print(predictions_few_shot[120])
print(predictions_few_shot_noins[120])
```

# Conclusão

O melhor desempenho foi obtido com aprendizado zero-shot, com uma acurácia de 91,6%. 

Isso é contra-intuitivo, pois era de se esperar que algum treinamento, ou seja, few-shot, traria melhor resultado.

Uma possível explicação é que para treinamento foram utilizados 5 textos reais do dataset, que por serem um pouco longos, acabam confundindo a máquina. Exemplos mais sucintos devem trazer melhores resultados. **Sugestão de trabalho futuro.**

O uso de few-shots permitiu, em geral, que a resposta fosse mais direta, como desejávamos, ou seja, simplesmente respondeu negativo ou positivo. No zero-shot, o formato da resposta saiu menos padronizado, exigindo mais tratamento. 

Só consideramos respostas que continham positivo ou negativo. Respostas neutras, respostas vagas ou respostas que exigiam uma interpretação foram consideradas erradas.

Assim, enfim, apesar do few-shot trazer respostas em geral mais diretas, essas respostas foram, na média, piores que as do zero-shot com um tratamento básico.

Outro resultado observado foi que o emprego de instrução, tanto no zero-shot como no few-shot, levou a um melhor desempenho: 91,6% e 90,2% com instrução, contra 82,5% no few-shot sem instrução, só com 5 exemplos.
